```python
# Joining all the strings into one and lowercasing it
text = ''.join(text).lower()
text
```

```python
print(counts)
```

```python
text = ''.join(filter(lambda x: counts[x] > 30, text)) # фильтруем редкие буквы такие как 'э' или 'w'
tokens = list(set(text))
print(f'number of characters: {len(tokens)}')
token_to_id = {el: i for i, el in enumerate(tokens)}
print(token_to_id)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
data = torch.tensor(tuple(map(token_to_id.get, text)))
print(data)
data.shape
```

```python
def get_batch(data, batch_size=200):
    ind = np.random.randint(0, data.shape[0] - batch_size)
    return data[ind: ind + batch_size]
```

```python
def encode(text):
   return list(map(token_to_id.get, text))
  
def decode(data):
    return list(map(lambda x: tokens[x], data.tolist()))
```

```python
batch = get_batch(data, 200)
print(batch)
print(''.join(decode(batch)))
```

```python
class RNNLoop(nn.Module):
    def __init__(self, num_tokens=len(tokens), emb_size=200, num_units=128):
        super(self.__class__, self).__init__()
        self.emb = nn.Embedding(num_tokens, emb_size)
        self.rnn = nn.LSTM(emb_size, num_units, batch_first=True, num_layers=2)
        self.hid_to_logits = nn.Linear(num_units, num_tokens)
    
    def forward(self, X):
        embedding = self.emb(X)
        h_seq, _ = self.rnn(embedding)
        next_logits = self.hid_to_logits(h_seq)
        # return next_logits
        next_logp = F.log_softmax(next_logits, dim=-1)
        return next_logp

model = RNNLoop(emb_size=32, num_units=128,)
opt = torch.optim.AdamW(model.parameters())
history = []
```

```python
from IPython.display import clear_output


for i in range(2000):
    batch_ix = get_batch(data)
    x = batch_ix[:-1]
    y = batch_ix[1:]

    predicted = model(x)
    loss = F.nll_loss(predicted, y)
    # loss = F.cross_entropy(predicted, y)
    opt.zero_grad()
    loss.backward()
    opt.step()
    history.append(loss.item())
    if (i + 1) % 50 == 0:
        clear_output(True)
        plt.plot(history,label='loss')
        plt.legend()
        plt.show()
```

```python
def generate_sample(model, seed='реве', length=200, temperature=1.0):
      x = torch.tensor([encode(seed)])
      model(x[:, :-1])
      for i in range(length - len(seed)):
          out = model(x[:, -1])
          p_next = F.softmax(out / temperature, dim=-1).data.numpy()[0]
          # p_next = out.data.view(-1).div(temperature).exp()
          # next_ix = torch.multinomial(p_next, 1)[0]
          # next_ix = p_next.argmax()
          next_ix = np.random.choice(len(tokens), p=p_next)
          x = torch.cat([x, torch.tensor([[next_ix]])], dim=1)
      
      return ''.join(decode(x[0]))
```

```python
print(generate_sample(model, 'реве ', temperature=0.6))
```
